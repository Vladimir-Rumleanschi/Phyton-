```python
#Lösung mit Höhe des Baumes muss bei jedemKnoten zunächst auf 1 gesetzt werden.

class TreeNode(object): 
    def __init__(self, val): 
        self.val = val 
        self.left = None
        self.right = None
        self.height = 1
        
    def display(self):
        lines, *_ = self._display_aux()
        for line in lines:
            print(line)

    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.val
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.left._display_aux()
        right, m, q, y = self.right._display_aux()
        s = '%s' % self.val
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    def insert(self, root, key): 
      
        # binäre Suche
        if not root: 
            return TreeNode(key)
        elif key == root.val:
            return root
        elif key < root.val: 
            root.left = self.insert(root.left, key) 
        else: 
            root.right = self.insert(root.right, key) 
  
        # Höhe etzen
        root.height = 1 + max(self.getHeight(root.left), 
                           self.getHeight(root.right)) 
        # balance factor 
        balance = self.getBalance(root) 
        # Step 4 - If the node is unbalanced,  
        # then try out the 4 cases 
        if balance > 1: 
            #root.display()
            if self.getBalance(root.right) < 0:
                root.right = self.rightRotate(root.right) 
            return self.leftRotate(root)
  
        if balance < -1: 
            #root.display()
            if self.getBalance(root.left) > 0:
                root.left = self.leftRotate(root.left) 
            return self.rightRotate(root)
    
        return root 
  
    def leftRotate(self, z): 
  
        y = z.right 
        T2 = y.left 

        y.left = z 
        z.right = T2 
        z.height = 1 + max(self.getHeight(z.left), 
                         self.getHeight(z.right)) 
        y.height = 1 + max(self.getHeight(y.left), 
                         self.getHeight(y.right)) 

        return y 
  
    def rightRotate(self, z): 
  
        y = z.left 
        T3 = y.right 

        y.right = z 
        z.left = T3 

        z.height = 1 + max(self.getHeight(z.left), 
                        self.getHeight(z.right)) 
        y.height = 1 + max(self.getHeight(y.left), 
                        self.getHeight(y.right)) 
  
        # Return the new root 
        return y 
  
    def getHeight(self, root): 
        if not root: 
            return 0
  
        return root.height 
    
    def getTreeHeight(self, root): 
        if not root: 
            return 0
        
        return root.height 
  
    def getBalance(self, root): 
        if not root: 
            return 0
        return self.getHeight(root.right) - self.getHeight(root.left) 
        #return maxright - maxleft

root = TreeNode(74)
#, 42, 13, 55, 5, 23, 69, 17
for key in [94, 82,42, 13, 55, 5, 23, 69, 17]:
    root = root.insert(root, key)

root = TreeNode(74)
for key in [94, 82]:
    root = root.insert(root, key)
root.display()
```

```python
#Lösung 2 mit live traversierung

class TreeNode(object): 
    def __init__(self, val): 
        self.val = val 
        self.left = None
        self.right = None
        
    def display(self):
        lines, *_ = self._display_aux()
        for line in lines:
            print(line)

    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.val
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.left._display_aux()
        right, m, q, y = self.right._display_aux()
        s = '%s' % self.val
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    def insert(self, root, key): 
      
        # binäre Suche
        if not root: 
            return TreeNode(key)
        elif key == root.val:
            return root
        elif key < root.val: 
            root.left = self.insert(root.left, key) 
        else: 
            root.right = self.insert(root.right, key) 
  

        # balance factor 
        balance = self.getBalance(root) 
        print(root.val ,'balance:', balance)
        # Step 4 - If the node is unbalanced,  
        # then try out the 4 cases 
        if balance > 1: 
            #root.display()
            if self.getBalance(root.right) < 0:
                root.right = self.rightRotate(root.right) 
            return self.leftRotate(root)
  
        if balance < -1: 
            #root.display()
            if self.getBalance(root.left) > 0:
                root.left = self.leftRotate(root.left) 
            return self.rightRotate(root)
    
        return root 
  
    def leftRotate(self, z): 
  
        y = z.right 
        T2 = y.left 

        y.left = z 
        z.right = T2 

        return y 
  
    def rightRotate(self, z): 
  
        y = z.left 
        T3 = y.right 

        y.right = z 
        z.left = T3 

        # Return the new root 
        return y 
  
    def getBalance(self, root): 
        if not root: 
            return 0
        maxleft = 0
        maxright = 0
        if root.left:
            maxleft = self.inOrder(root.left)
 #           print('left ',maxleft)
            
        if root.right:
            maxright = self.inOrder(root.right)
    #        print('right ', maxright)
        return maxright - maxleft
    def inOrder(self,root):
        output = []
        stack = []
        seen = []
        stack.append(root)
        depth = 0
        maxdepth = 0
        while stack:
            node = stack.pop()
            if node not in seen and (node.left or node.right):
                if node.left:
                    stack.append(node.left)
                stack.append(node)
                depth = depth+1
                if  node.right:
                    stack.append(node.right)

                seen.append(node)
            else:
                maxdepth = max(maxdepth,depth+1)
                if node==root:
                    depth = 1
        return maxdepth

root = TreeNode(74)
#, 42, 13, 55, 5, 23, 69, 17
for key in [94, 82,42, 13, 55, 5, 23, 69, 17]:
    root = root.insert(root, key)
root.display()
root = TreeNode(74)
for key in [74, 94, 82,42, 13, 55]:
    print('insert ', key,' to ', root.val )
    root = root.insert(root, key)
    root.display()
    
root.display()
```
